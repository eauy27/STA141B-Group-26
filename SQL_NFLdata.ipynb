```python
import pandas as pd
import sqlite3 as sql
import openpyxl

avg_stats = ['AVG_QB','AVG_WR','AVG_TE','AVG_RB','AVG_DE','AVG_LB','AVG_OLB','AVG_ILB','AVG_CB','AVG_DB','AVG_S','AVG_FS','AVG_DT','AVG_K'] # List of average stats for each NFL position
total_stats = ['TOT_QB','TOT_WR','TOT_TE','TOT_RB','TOT_DE','TOT_LB','TOT_OLB','TOT_ILB','TOT_CB','TOT_DB','TOT_S','TOT_FS','TOT_DT','TOT_K'] # List of total stats """"""
```

```python
db = sql.connect('nfl_draft_analysis.db') # Establish Database connection

xlsx = pd.ExcelFile('nfl_draft_analysis.xlsx') # Load Excel file
for sheet in avg_stats + total_stats:
    
    df = pd.read_excel(xlsx, sheet_name = sheet) # Creates a Data Frame from each Excel sheet
    df.to_sql(sheet, db, index=False, if_exists='replace') # Pushes each Data Frame as a table in the Database

df_master = pd.read_excel(xlsx,sheet_name = 'MASTER_all_players') # Adds the Master table with all stats and season
df_master.to_sql('master', db, index = False, if_exists='replace')
```

Everything above this sets up the DataBase connection and tables

```python
#### Things to do using SQL
# take average position stats (the important ones)
# take average career stats of each player (the ones that are most important to that position)
# compare the player stats to the average
# for stats that arent important to their position, N/A
```

Offensive positions

```python
#QB
query_QB = """
SELECT
    t.player_name_full,
    t.position,
    t.college,
    t.draft_slot,
    t.round,
    t.pick,

    t.completions,
    a.completions_mean,
    t.completions - a.completions_mean AS completions_diff,

    t.attempts,
    a.attempts_mean,
    t.attempts - a.attempts_mean AS attempts_diff,

    t.passing_yards,
    a.passing_yards_mean,
    t.passing_yards - a.passing_yards_mean AS passing_yards_diff,

    t.passing_tds,
    a.passing_tds_mean,
    t.passing_tds - a.passing_tds_mean AS passing_tds_diff,

    t.passing_interceptions,
    a.passing_interceptions_mean,
    t.passing_interceptions - a.passing_interceptions_mean AS passing_interceptions_diff

FROM TOT_QB t
JOIN AVG_QB a
ON t.draft_slot = a.draft_slot
"""

df_qb = pd.read_sql(query_QB, db)

df_qb.head()
```

```python
# DB 
query = """ SELECT t.player_name_full, 
                    t.college,
                    t.draft_slot,
                    t.round,
                    t.pick,

                    m.season,

                    -- TOTAL TACKLES
                    t.def_tackles_solo, 

                    -- TOTAL INTERCEPTIONS
                    t.def_interceptions,

                    -- TOTAL PASSES DEFENDED
                    t.def_pass_defended,

                    -- TOTAL SACKS
                    t.def_sacks,

                    -- TOTAL TACKLES FOR LOSS
                    t.def_tackles_for_loss

            FROM TOT_DB t
            JOIN master m ON m.player_id = t.player_id

            GROUP BY t.player_name_full
        """

df_DB = pd.read_sql(query,db)
df_DB.head()
```
