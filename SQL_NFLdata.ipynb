```python
import pandas as pd
import sqlite3 as sql

df = pd.read_csv('nfl_master_table.csv')
db = sql.connect(':memory:')

df.to_sql('players', db, index=False, if_exists='replace') # Pushes DataFrame as 'players' table into Database
```

```python
# group by position and draft round
query = """
        SELECT name_key as name, position, season, round
        FROM players
        WHERE season >= 2016
        GROUP BY name, position, round, season
        """
clean_table = pd.read_sql(query,db)
```
