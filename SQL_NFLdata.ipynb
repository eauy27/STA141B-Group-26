```python
# Count for number of players of a specific school that perform above average with respect to their draft position and position on the field
```

```python
import pandas as pd

# define averages
avg_QB = pd.read_sql("SELECT * FROM AVG_QB", db)
avg_RB = pd.read_sql("SELECT * FROM AVG_RB", db)
avg_WR = pd.read_sql("SELECT * FROM AVG_WR", db)
avg_TE = pd.read_sql("SELECT * FROM AVG_TE", db)
avg_OLB = pd.read_sql("SELECT * FROM AVG_OLB", db)
avg_K = pd.read_sql("SELECT * FROM AVG_K", db)
avg_FS = pd.read_sql("SELECT * FROM AVG_FS", db)
avg_DB = pd.read_sql("SELECT * FROM AVG_DB", db)
avg_CB = pd.read_sql("SELECT * FROM AVG_CB", db)
avg_S = pd.read_sql("SELECT * FROM AVG_S", db)
avg_DT = pd.read_sql("SELECT * FROM AVG_DT", db)
avg_DE = pd.read_sql("SELECT * FROM AVG_DE", db)
avg_ILB = pd.read_sql("SELECT * FROM AVG_ILB", db)
avg_LB = pd.read_sql("SELECT * FROM AVG_LB", db)

# dictionary of player dataframes
player_dfs = {
    "QB": df_QB,
    "RB": df_RB,
    "WR": df_WR,
    "TE": df_TE,
    "OLB": df_OLB,
    "K": df_K,
    "FS": df_FS,
    "DB": df_DB,
    "CB": df_CB,
    "S": df_S,
    "DT": df_DT,
    "DE": df_DE,
    "ILB": df_ILB,
    "LB": df_LB
}

# dictionary of average dataframes
avg_dfs = {
    "QB": avg_QB,
    "RB": avg_RB,
    "WR": avg_WR,
    "TE": avg_TE,
    "OLB": avg_OLB,
    "K": avg_K,
    "FS": avg_FS,
    "DB": avg_DB,
    "CB": avg_CB,
    "S": avg_S,
    "DT": avg_DT,
    "DE": avg_DE,
    "ILB": avg_ILB,
    "LB": avg_LB
}

# significant stats for each position
stats_dict = {
    "QB": ["completions", "attempts", "passing_yards", "passing_tds", "passing_interceptions"],
    "RB": ["carries", "rushing_yards", "rushing_tds", "receptions", "targets"],
    "WR": ["receptions", "targets", "receiving_yards", "receiving_tds", "receiving_air_yards"],
    "TE": ["receptions", "targets", "receiving_yards", "receiving_tds", "receiving_air_yards"],
    "OLB": ["def_tackles_solo", "def_sacks", "def_qb_hits", "def_interceptions", "def_tackles_for_loss"],
    "K": ["fg_made", "fg_att", "fg_missed", "fg_long", "pat_made"],
    "FS": ["def_tackles_solo", "def_interceptions", "def_pass_defended", "def_tackles_for_loss", "def_qb_hits"],
    "DB": ["def_tackles_solo", "def_interceptions", "def_pass_defended", "def_sacks", "def_tackles_for_loss"],
    "CB": ["def_tackles_solo", "def_interceptions", "def_pass_defended", "def_sacks", "def_tackles_for_loss"],
    "S": ["def_tackles_solo", "def_interceptions", "def_pass_defended", "def_tackles_for_loss", "def_qb_hits"],
    "DT": ["def_tackles_solo", "def_sacks", "def_qb_hits", "def_tackles_for_loss", "def_fumbles_forced"],
    "DE": ["def_tackles_solo", "def_sacks", "def_qb_hits", "def_tackles_for_loss", "def_fumbles_forced"],
    "ILB": ["def_tackles_solo", "def_sacks", "def_qb_hits", "def_interceptions", "def_tackles_for_loss", "def_pass_defended"],
    "LB": ["def_tackles_solo", "def_sacks", "def_qb_hits", "def_interceptions", "def_tackles_for_loss", "def_pass_defended"]
}

# function to compare player-season rows to the averages
def compare_players_to_avg(player_df, avg_df, stats, position_name):
    df = player_df.copy()
    avg = avg_df.copy()

    # keep only draft_slot and the mean columns from the average table
    avg_cols = ["draft_slot"] + [stat + "_mean" for stat in stats]
    avg = avg[avg_cols]

    # merge player rows with averages by draft_slot
    df = pd.merge(df, avg, on="draft_slot", how="left")

    # compare each stat to its respective average
    flag_cols = []

    for stat in stats:
        mean_col = stat + "_mean"
        flag_col = stat + "_above"

        # fewer missed goals is better so we are looking for a lower value
        if stat == "fg_missed":
            df[flag_col] = (df[stat] < df[mean_col]).astype(int)
        else:
            df[flag_col] = (df[stat] > df[mean_col]).astype(int)

        flag_cols.append(flag_col)

    # count how many stats are above average
    df["num_above_stats"] = df[flag_cols].sum(axis=1)

    # count players that are above average as "1", otherwise "0" (players are above average if at least 3 statistics are above average)
    df["above_avg_player"] = (df["num_above_stats"] >= 3).astype(int)

    df["position"] = position_name

    return df[[
        "player_name_full",
        "college",
        "season",
        "position",
        "draft_slot",
        "num_above_stats",
        "above_avg_player"
    ]]

# store results
all_results = []

for pos in player_dfs:
    result = compare_players_to_avg(
        player_dfs[pos],
        avg_dfs[pos],
        stats_dict[pos],
        pos
    )
    all_results.append(result)

all_players = pd.concat(all_results, ignore_index=True)

# count number of above average players by school for each season
school_year_counts = (
    all_players.groupby(["season", "college"])["above_avg_player"]
    .sum()
    .reset_index()
    .rename(columns={"above_avg_player": "num_players_above_avg"})
)

# sort by season then by number of above average players in descending order
school_year_counts = school_year_counts.sort_values(
    ["season", "num_players_above_avg"],
    ascending=[True, False]
)

# keep top 10 schools for each season
top10_by_season = school_year_counts.groupby("season").head(10).copy()

top10_by_season = top10_by_season.sort_values(
    ["season", "num_players_above_avg"],
    ascending=[True, False]
)

# enter year to see top 10 schools for that year
def top10_for_year(year):
    return top10_by_season[top10_by_season["season"] == year]
# example
top10_for_year(2015)
```
